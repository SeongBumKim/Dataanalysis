# 결측치

결측치는 **비어있는 데이터**를 의미합니다.

결측치에 대한 처리는 매우 중요합니다.

결측치에 대한 처리를 해주려면 **다음의 내용**을 반드시 알아야 합니다.

1. 결측 데이터 확인
2. 결측치가 **아닌** 데이터 확인
3. 결측 데이터 **채우기**
- 시계열 : 근처의 값으로 채우기, 추세, 계절성(단기적, 패턴이 일정), 주기(cycle, 장기적인, 기간이 일정치 않은)를 이용하여 판단
- 정규분포의 데이터 : 평균, 중앙값, 최빈값 등의 대표값으로 채우기
- 데이터의 특성을 고려해서 값을 정함 (틀릴 가능성 높음)
4. 결측 데이터 **제거하기**
- 데이터의 왜곡을 방지할 수 있음
- 수집비용, 다른 데이터 손실, 편향 데이터 등의 여러 가지 문제가 존재함

5. 결측 데이터를 예측하는 모델을 만드는 것

## 결측치 채우기 - fillna()

`fillna()`를 활용하면 결측치에 대하여 **일괄적으로 값을 채울 수** 있습니다.

```python
# 원본을 copy하여 df1 이름 붙이기
df0 = df.copy()
df1 = df.copy()
```

888번 index의 **결측치가 700으로 채워**진 것을 확인할 수 있습니다.

```python
# df1의 'age' 컬럼의 결측치를 700으로 채우기하여, 마지막 5개 데이터를 출력해 봅니다.
df0['age'].fillna(700).tail()
```

```python
# df1의 'age' 컬럼을 결측치가 700으로 채워진 내용으로 갱신합니다.
df0['age'] = df0['age'].fillna(700)
df0['age'].tail()
```

**카테고리 형 데이터**을 채워주기 위해서는 다음과 같은 과정을 거쳐야 합니다.

이미 카테고리가 추가된 'A'나 'B'는 바로 fillna() 할 수 있습니다.

```python
# df1의 'deck' 컬럼의 결측치를 'A'로 채우기합니다.  (갱신은 하지 않음)
df1['deck'].fillna('A').tail()
```

하지만, 없는 카테고리로 채워주고자 할 때는 먼저 `cat.add_categories`로 카테고리를 추가한 후 채워야 합니다.

```python
# cat은 category의 지정자
# add_categories (카테고리 추가)
# df1의 'deck'의 cat.add_categories()를 사용하여 'No Data'라는 카테고리를 추가하고
# 결측치를 'No Data'로 채우기 합니다.
# df1['deck'].cat.add_categories('No Data')
df1['deck'] = df1['deck'].fillna('No Data')
df1['deck'].tail()
```

## NaN 값이 있는 데이터 제거하기 (dropna)
- DataFrame.dropna(axis=0, how='any', thresh=None, subset=None, inplace=False)
- axis=0/1 : 삭제할 것이 행/열
- how={'any','all'} => 'any': NaN이 하나라도 있으면 제거, 'all' : 모든 값이 NaN 이면 제거
- thresh=정수값 => NaN이 아닌 데이터가 정수 미만이면 제거
- subset=[컬럼목록/행목록] => na값 확인하는 컬럼/행 목록

```python
temp = pd.DataFrame([[np.nan, 1, 2, 3], [np.nan, 1, 2, np.nan],
                     [np.nan, np.nan, 2, np.nan], [np.nan, 1, np.nan, 3]])
temp
```

```python
temp.dropna()
```

```python
temp.dropna(axis=1)
```

```python
temp.dropna(how='all')
```

```python
temp.dropna(axis=1, how='all')
```

```python
temp.dropna(axis=0, thresh=2)
```

```python
temp.dropna(axis=0, subset=[3])  # axis = 0 일때는 subset=[column_label]
```

```python
temp.dropna(axis=1, subset=[3])  # axis = 1 일때는 subset=[index_label]
```

```python
df1 = df.copy()
```

```python
df1.tail()
```

`dropna()`로 **1개 라도 NaN 값이 있는 행**은 제거할 수 있습니다. (`how='any'`)

```python
# df1에서 1개 라도 NaN 값이 있는 행을 제거합니다.
df1.dropna(axis=0, how='any')
```

기본 옵션 값은 `how=any`로 설정되어 있으며, 다음과 같이 변경할 수 있습니다.

- **any**: 1개 라도 NaN값이 존재시 drop
- **all**: 모두 NaN값이 존재시 drop

```python
# df1에서 모든 값이 결측치인 행을 제거합니다. (how='all' 사용)
df1.dropna(axis=0, how='all')
```

```python
df1.dropna(axis=0, thresh=14)
```
